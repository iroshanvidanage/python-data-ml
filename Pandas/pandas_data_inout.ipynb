#### Need to create a SQL engine in the memory

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///:memory:')
# a temp sql engine is now running in memory
```

```python
df.to_sql('temp_table', engine)
```

```python
sqldf = pd.read_sql('temp_table', con=engine)
sqldf
```
