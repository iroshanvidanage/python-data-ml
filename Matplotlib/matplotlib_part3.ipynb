## Plot appearance
### setting colors

```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])


# for colors we can specify the rgb color code or the basic color names
# line thickness linewidth or lw can be set
# line visibility 0-1
# linestyle or ls
# markers for points
ax.plot(x,y,color='green',lw=5,alpha=0.5,ls='-',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='red')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',lw=2,ls='-')

# upperbounds and lower bounds for axes
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
