```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

# this will clear out any overlapping
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
value = 1
for current in axes:
    current.plot(x * value, y / value)
    value += 1
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Figure size, Aspect ratio & DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### how to save a figure

```python
fig.savefig(fname='my_fig.png',dpi=200)
```

### how to add a legend

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')

# location can be specified
# a location map is defined or the position can be specified as a tuple
ax.legend(loc=0)
```
